# 1. Introducción y Objetivo

La paleobiología es la ciencia encargada de la historia de la vida terrestre mediante el análisis de fósiles y evidencias geológicas asociadas (Cracraft, 1981). Su estudio es esencial para entender la vida en el pasado y la evolución de la biodiversidad en el tiempo geológico, permitiendo a los científicos conocer los distintos tipos de fauna que habitaban el planeta, ambientes de localización, distribución geográfica, composición, forma y anatomía, análisis de dieta y hábitos de vida (Prothero, 2013). Además, de proporcionar evidencias sobre cambios climáticos, y patrones de extinción y supervivencia de los organismos (Fordham et al., 2020), con aplicaciones en la predicción de cambios ambientales actuales y futuros (Yasuhara & Deutsch, 2022), así como la conservación de ecosistemas (Barnosky et al., 2017). 
Por las razones anteriores, el análisis de datos paleobiológicos es fundamental. Y especialmente, el análisis de grandes volúmenes de estos datos es relevante para descubrir patrones, realizar asociaciones y obtener conclusiones más certeras, que puedan aportar en el avance paleobiológico. En este contexto, la manipulación y análisis de datos utilizando herramientas informáticas se han convertido en habilidades esenciales para los paleobiólogos modernos (Sepkoski & Ruse, 2009). Así mismo, la creciente disponibilidad de conjuntos de datos ha hecho que la codificación sea una parte integral de la investigación paleobiológica, resultando especialmente útiles herramientas de programación como R y Python por sus ventajas para el manejo de grandes datos, y la realización de análisis categóricos, matemáticos y estadísticos (Colliau et al., 2017). Un ejemplo de esto es el Paquete Palaeoverse de R, una herramienta creada para ayudar en la preparación y exploración de datos para la investigación paleobiológica cuantitativa (Jones et al., 2023).

En la actualidad, Python sobresale como uno de los lenguajes de programación más utilizados en el mundo por su simplicidad y versatilidad, pues ofrece una variedad de librerías y herramientas que facilitan el procesamiento y análisis de datos geocientíficos como los de tipo paleobiológico (Gaurav & Sindhu, 2018; Rawat, 2020). Desde la limpieza y organización de grandes conjuntos de datos fósiles, la realización de análisis exploratorios, hasta la visualización gráfica de datos de interés. Siendo utilizado, por ejemplo, en estudios de tendencias de investigación (Wilson et al., 2018) y automatización de identificación taxonómica basada en imágenes de fósiles (Liu et al., 2023). Este lenguaje de programación permite a los investigadores realizar análisis complejos de forma eficiente y reproducible.

El objetivo de este trabajo es analizar los registros de la base de datos paleobiológicos de la Antártida, mediante el uso de herramientas y librerías de Python, para la afinación de representaciones gráficas que permitan la distinción de aspectos importantes sobre el pasado geológico, climático y biológico de cualquier región y sus cambios a lo largo de millones de años.  La base de datos está enfocada en el continente Antártico, una zona de gran interés en la investigación paleobiológica que, por su aislamiento geográfico, baja perturbación y alteración antrópica, han permitido una conservación óptima de los restos fósiles y rasgos geológicos, ofreciendo una ventana valiosa para el estudio de la vida en el pasado.

### Exploración automatizada de datos --- Obtención de Ambiente Taxonómico y Dieta dado el Nombre Aceptado de una ocurrencia fósil


Se creó la función **taxon_and_diet(df, accepted_name)** como un recurso para la obtención de datos de interés como el ambiente taxonómico y la dieta de una ocurrencia fósil a partir de su nombre común (*Accepted name*). Como ejemplo se tomó a la especie Solemya.

```python
def taxon_and_diet(df, accepted_name):
    result = df[df["Accepted Name"] == accepted_name][["Taxon Environment","Diet"]]
    if result.empty:
        return None
    else:
        return result.iloc[0].to_dict()


accepted_name = "Solemya"
taxon_diet = taxon_and_diet(c_antar, accepted_name)
if taxon_diet:
    print(accepted_name)
    print(f"-Taxon Environment: {taxon_diet["Taxon Environment"]}\n-Diet: {taxon_diet["Diet"]}")
else:
    print("El nombre que intenta buscar no se encuentra en el DataFrame")

```

### Diagrama de Barras - Número de especies por lista de Phylums

Se creó la función BarSpecies que toma dos argumentos, el primero es el dataframe filtrado mientras el segundo corresponde a la lista de los phylum que deseamos explorar. Los argumentos sirven para dibujar un diagrama de barras que provee un conteo de las especies encontradas según un phylum determinado. Al explorar la cantidad de especies de un phylum, por ejemplo, al explorar *Bryozoa*, predominan **Cellarinella** cuyos ambientes preferidos son de temperaturas bajas y profundas, mientras **Cellaria** se encuentran en ambientes de aguas someras a moderadamente profundas y semi-templadas. De igual forma, al analizar *Arthropoda* se encuentran considerables ocurrencias de **Hoploparia stokesi**, debido a que es un tipo extinto de langosta puede dar indicios de tolerancia a la salinidad, identicando condiciones marinas y de circulación oceánica.

```python
def BarSpecies(dataframe, lista_phyla):
    fig, axes = plt.subplots(nrows=len(lista_phyla), figsize=(10, 10 * len(lista_phyla)))
    
    for i, phylum in enumerate(lista_phyla):
        # Filtrar el DataFrame por el phylum actual
        df_filtrado = dataframe[dataframe['Phylum'] == phylum]
        
        # Contar el número de especies por phylum
        conteo_especies = df_filtrado['Accepted Name'].value_counts()
        
        # Crear el gráfico de barras para el phylum actual
        ax = axes[i] if len(lista_phyla) > 1 else axes
        conteo_especies.plot(kind='bar', ax=ax)
        ax.set_xlabel('Especies')
        ax.set_ylabel('Número de especies')
        ax.set_title(f'Número de especies para {phylum}')
        ax.tick_params(axis='x', rotation=90)
    
    plt.tight_layout()
    plt.show()


BarSpecies(c_antar,["Bryozoa","Arthropoda"])
```

### Diagrama apilado - Distribución por clase y orden según el Phylum


Se creó la función **Phylum_class_order** que recibe los datos filtrados y entre comillas el phylum, con el objetivo de hacer un diagrama de barras apilado por clase y orden, correspondientes a los niveles inferiores de la taxonomía dada en la base. Así el orden corresponde a Phylum > Clase (*class*) > Orden (*order*). Al aplicarlo sobre el phylum Myzozoa se observa una gran cantidad de individuos del orden **Gonyaulacales** ya que forman parte del fitoplancton esencial para la cadena alimentaria marina.

```python
def Phylum_class_order(rawdata,phylum):
    data = rawdata.loc[rawdata["Phylum"] == phylum, ["Class", "Order"]]
    grouped = data.groupby(['Class', 'Order']).value_counts().unstack()
    print(grouped)
    plt.figure(figsize=(25, 6)) 
    grouped.plot(kind='bar', stacked=True)
    plt.xlabel('Classes')
    plt.ylabel('Number of occurrences')
    plt.title(f"Distribución por clases y órdenes en el Phylum {phylum}")
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
    plt.show()

Phylum_class_order(c_antar,"Myzozoa")
```

### Distribución de las edades mínimas por tipo de individuo 

Se hace un filtrado de los individuos del phylum **Sphenophyta** y se toman las columnas "Accepted Name","Minimum Age","Maximum Age", y "Reference Author".

```python
spheno=c_antar.loc[(c_antar["Phylum"]=="Sphenophyta"),
["Accepted Name","Minimum Age","Maximum Age","Reference Author"]]
spheno
```

Luego se crea la función **BoxplotAges** para crear un boxplot en base a las edades mínimas por nombre común de especie *(Accepted name)*. Donde se puede inferir que entre más irregular es el gráfico, menos diversidad de edades fue encontrado, debido a que los autores reportan más de un dato de una misma especie con la misma edad mínima.

```python
def BoxplotAges(data):
    # Define objects
    fig, ax = plt.subplots(figsize=(6,6))
    sns.boxplot(x="Accepted Name", y="Minimum Age", data=data, ax=ax)
    ax.tick_params(axis="x", rotation=45)

    plt.show()
BoxplotAges(spheno)
```

### Rango de edades por especie según autores

Para ilustrar el rango de edades por especie según los diferentes autores de referencia tomamos como ejemplo a la especie **Paracalamites** del phylum *Sphenophyta*. Determinando una resiliencia de la especie en eventos catastróficos como la extinción masiva del Pérmico-Triásico.

### Rango de edades por especie según autores

Finalmente, para obtener un rango de edades de cualquier especie de la database, se creó una función llamada **SpecieAgeRange** la cual toma la data, y un string con el nombre de la especie, y dibuja un gráfico de dispersión del rango de edades de acuerdo a lo encontrado por varios autores. Por ejemplo, los datos de la especie ***Cucullaea*** pueden inferir nuevos ambientes de transición entre ecosistemas marinos y terrestres en edades recientes.

```python
def SpecieAgeRange(data, species_name):
    species_data = data[(data["Accepted Name"] == species_name)]
    
    if species_data.empty:
        print(f"No se encontraron datos para la especie {species_name}.")
        return
    authors = np.array(species_data["Reference Author"])
    min_age = np.array(species_data["Minimum Age"])
    max_age = np.array(species_data["Maximum Age"])
    fig, ax = plt.subplots(figsize=(9, 6))
    ax.plot(authors, min_age, marker="+", ms=14, color="darkblue", label="Minimum Age")
    ax.plot(authors, max_age, marker="+", ms=14, color="red", label="Maximum Age")
    ax.set_xlabel("Autores")
    ax.set_ylabel("Edad en Millones de años (Ma)")
    ax.set_title(f"Rango de edades de la especie {species_name} según varios autores")
    ax.axhline(y=np.mean(min_age + max_age), c="k", ls="--") 
    ax.axhline(y=np.min(min_age), c="k", ls="--", alpha=0.5)  
    ax.axhline(y=np.max(max_age), c="k", ls="--", alpha=0.5)  
    ax.set_ylim(np.min(min_age) - 5, np.max(max_age) + 5)
    ax.fill_between(authors, min_age, max_age, color="lightblue", alpha=0.1)
    ax.legend()
    plt.tight_layout()
    plt.show()


SpecieAgeRange(c_antar, "Cucullaea")
```
